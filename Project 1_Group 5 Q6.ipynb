```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
os.listdir()
```

```python
df_house = pd.read_excel("House_prices _region_2010_2020.xlsx")
```

```python
#!pip3 install openpyxl
```

```python
df_house.head()
```

```python
# melt the dataframe
df_house_melt = pd.melt(df_house, id_vars=["Region"], var_name="Month", value_name="Price")
```

```python
df_house_melt
```

```python
df_house.set_index("Region", inplace=True)
```

```python
# use fivethirtyeight style for matplotlib
plt.style.use("fivethirtyeight")
```

```python
df_house_melt.head()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x="Month", y="Price", data=df_house_melt, hue="Region")
plt.xticks(rotation=45)
plt.title("House prices in 2010-2020")
```
