```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Final_Clean V2.csv")
```

```python
df2 = pd.read_csv("Rental_Cost_Cleaned_v2.csv")
```

```python
df1.head()
```

```python
df1.info()
```

```python
str_columns = df1.select_dtypes('object').columns
```

```python
str_columns = str_columns[1:]
```

```python
for col in str_columns:
    df1[col] = df1[col].str.replace('$','')
    df1[col] = df1[col].str.replace(',','')
    df1[col] = pd.to_numeric(df1[col], errors='coerce')
```

# Find top 10 suburbs that has highest increase in price percentage from 2010 to 2020

```python
df1.head()
```

```python
# pct change = (new - old)/new
```

```python
df1["percentage_change"] = (df1["Median 2020"] - df1["Median 2010"])/df1["Median 2020"]
```

```python
df1.head()
```

```python
df1.sort_values("percentage_change",ascending=False, inplace=True)
```

```python
df1.head(2)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df1.index[:10],df1.iloc[:10,-1] * 100)
plt.xticks(rotation=45)
plt.title("Top 10 suburbs highest increase in housing price over 10 years")
plt.savefig("top10housing.png")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df1.index[-10:],df1.iloc[-10:,-1] * 100)
plt.xticks(rotation=45)
plt.title("Top 10 suburbs Lowest increase in housing price over 10 years")
plt.savefig("bottom10housing.png")
```

```python
df2["percentage_change"] = (df2["Total Median 2020"] - df2["Total Median 2010"])/df2["Total Median 2020"]
```

```python
df2.sort_values("percentage_change",ascending=False, inplace=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df2.index[:10],df2.iloc[:10,-1] * 100)
plt.xticks(rotation=90)
plt.title("Top 10 suburbs highest increase in renting price over 10 years")
plt.savefig("top10renting.png")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df2.index[-10:],df2.iloc[-10:,-1] * 100)
plt.xticks(rotation=45)
plt.title("Top 10 suburbs Lowest increase in renting price over 10 years")
plt.savefig("bottom10renting.png")
```
